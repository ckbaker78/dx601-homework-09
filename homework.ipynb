### Problem 16

Rescale the vectors $[2, 0, 0, 0]$ and $[0, 1, 1, 1]$ so that they are orthonormal.

Save the first rescaled vector in `p16a` and the second rescaled vector in `p16b`.

```python
# YOUR CHANGES HERE

# orthronormal = perpendicular AND magnitude = 1 

a = np.array([2,0,0,0])

b1 = np.array([0,1,1,1]) 

p16a = (a) / np.linalg.norm(a) # this divides a by its magnitude, rescaling the size to one
```

```python
p16a
```

```python
# YOUR CHANGES HERE

b2 = b1 - np.dot(p16a,b1)  * p16a # this is subtracting off the b1 componet that points in the direction of p16 allowing for orthogonal 


p16b = b2 / np.linalg.norm(b2) # now p16a and p16b are orthonormal to one another 

b2 
```

### Problem 18

Set `p18` to the angle $\theta$ between the vectors in `u18` and `v18`.
Use the measure of the angle in radians, not degrees.

Hint: you can use the cosine similarity to calculate $\mathrm{cos}~\theta$.
The inverse cosine function to recover the measure of the angle is often called the arc cosine, and both the math and NumPy modules use this for their naming.

```python
# DO NOT CHANGE

u18 = np.asarray([0.3, 0.2, 0.9, -0.4])
u18
```

```python
# DO NOT CHANGE

v18 = np.asarray([-0.1, 0.3, 0.6, -0.2])
v18
```

```python
# YOUR CHANGES HERE

p18 = np.arccos(np.dot(u18,v18)/ (np.linalg.norm(v18) *np.linalg.norm(u18))) # arc cosine cancels out the cosine in the cosine similarity equation, leaving just the angle
```

```python
p18
```
